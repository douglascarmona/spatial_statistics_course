# <font color='darkgreen'> Sets o Conjuntos </font>

Los sets en Python son una estructura de datos usada para almacenar elementos de una manera similar a las listas, pero con algunas diferencias.

Debido a que los conjuntos no pueden tener múltiples ocurrencias del mismo elemento, hace que los conjuntos sean muy útiles para eliminar eficientemente los valores duplicados de una lista o tuple y realizar operaciones de matemáticas comunes como sindicatos e intersecciones.

Para crear un set en Python se puede hacer con <font color='darkred'>Sets o Conjuntos</font>
 y pasando como entrada cualquier tipo iterable, como puede ser una lista. Se puede ver como a pesar de pasar elementos duplicados como dos 8 y en un orden determinado, al imprimir el set no conserva ese orden y los duplicados se han eliminado.

```python
# Crear set Python

s = set([5, 4, 6, 8, 8, 1])
print(s)       
print(type(s)) 
```

### Operaciones con sets


A diferencia de las listas, en los set no podemos modificar un elemento a través de su índice. Si lo intentamos, tendremos un TypeError.

```python
s = set([5, 6, 7, 8]) # declaro el set
#s[0] = 3 # intento modificar el primer elemento
```

Los elementos de un set deben ser **inmutables**, por lo que un elemento de un set no puede ser ni un diccionario ni una lista. Si lo intentamos tendremos un TypeError

```python
lista = ["Perú", "Argentina"]
#s = set(["México", "España", lista])
```

**Los sets se pueden iterar de la misma forma que las listas.**

```python
s = set([5, 6, 7, 8])
for i in s:
    print(i) 
```

    
Con la función **len()** podemos saber <font color='darkcyan'>la longitud total del set</font>. Como ya hemos indicado, los duplicados son eliminados.



```python
s = set([1, 2, 2, 3, 4])
print(len(s)) 
```

También podemos saber si un elemento está presente en un set con el operador in. Si el valor existe en el set, se devolverá True.

```python
s = set(["Guitarra", "Bajo"])
print("Guitarra" in s) 
print("Piano" in s)
```

## <font color='darkgreen'> Métodos sets</font>



Los sets tienen además diferentes funcionalidades, que se pueden aplicar en forma de operador o de método. Por ejemplo, el operador | nos permite realizar la unión de dos sets, lo que equivale a juntarlos. El equivalente es el método union() que vemos a continuación.

```python
s1 = set([1, 2, 3]) # declaramos el primer set
s2 = set([3, 4, 5]) # declaramos el segundo set
print(s1 | s2) # unimos los dos sets y lo mostramos
```

**El método add() permite añadir un elemento al set.**

s.add(<elem>)

```python
set1 = set([1, 2])
set1.add(3)
print(set1)
```

**El método remove() elimina el elemento que se pasa como parámetro. 
Si no se encuentra, se lanza la excepción KeyError.**

 
s.remove(<elem>)

```python
set2 = set([1, 2,4,7])
set2.remove(2)
print(set2) 
```

**El método discard() es muy parecido al remove(), borra el elemento que se pasa como parámetro,y si no se encuentra no hace nada.**

s.discard(<elem>)

```python
set3 = set([1, 2, 5])
set3.discard(3)
print(set3)
```

**El método pop() elimina un elemento aleatorio del set.**

s.pop()

```python
set4 = set([1, 2, 9, "A"])
set4.pop()
print(set4) 
```

**El método clear() elimina todos los elementos de set.**

s.clear()

```python
set5 = set([1, 8, 6, "B"])
set5.clear()
print(set5)
```

Los sets cuentan con una gran cantidad de métodos que permiten realizar operaciones con dos o más, como la unión o la intersección.

Podemos calcular **la unión entre dos sets usando el método union()**. Esta operación representa la “mezcla” de ambos sets. Nótese que el método puede ser llamado con más parámetros de entrada, y su resultado será la unión de todos los sets

```python

s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.union(s2)) 
```

También podemos calcular **la intersección entre dos o más set**. Su resultado serán aquellos elementos que pertenecen a ambos sets.

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.intersection(s2)) 
```

# <font color="darkgreen">Slicing</font> 

La expresión slicing hace referencia a la operación por medio de la cual se extraen elementos de una secuencia, tal como una lista o una cadena de caracteres. Dependiendo del caso, los elementos podrían ser consecutivos o podrían estar separados dentro de la secuencia original. Dado que esta operación es de uso bastante frecuente, Python ofrece características que hacen posible construir slices de forma breve, sin tener que construir funciones especializadas para hacerlo

Supongamos que tenemos la siguiente cadena de caracteres con la que inicia el himno de Argentina:

```python
cadena = '¡Oid mortales el grito sagrado: Libertad,Libertad,Libertad!'
```

Cómo podemos hacer para extraer elementos de esta cadena usando paréntesis cuadrados y una posición numerada desde cero.

```python
cadena[0] # estamos extrayendo un slice de la cadena original con el primer caracter
```

```python
cadena[1:4] # extraemos los caracteres desde la posición 1 hasta la 4 (excluida), 
```

```python
cadena[30:] # recortamos de la posición 30 en adelante
```

```python
cadena[:30] # recortamos hasta la posición 30 exclusive
```

```python
cadena[None:30] # otra forma de hacer lo mismo que la anterior
```

```python
# Crear una copia de la cadena original
cadena2 = cadena[:]
cadena2
```

**Slicing intermedio: posiciones negativas**

Algo muy interesante en Python con respecto a los slices, es que las posiciones de inicio y de fin no necesariamente tienen que ser números positivos: si se usan números negativos, Python asume que se están referenciando las posiciones numeradas desde la última hasta la primera, empezando con -1

```python
cadena3= "HOLA, GENTE!"
cadena3[-1] # consultamos la última posición
```

```python
cadena3[-2] # consultamos la anteúltima posición
```

```python
cadena3[-10: -6] #¿qué hace esto?
```

### <span style="color: darkblue;">Para acceder a los elementos de una matriz o recorrerlos, podemos también utilizar funciones lógicas y obtener resultados "booleanos" (True or False). <span/>
Por ejemplo, podríamos preguntamos que elementos de la matriz son mayores a cero e inspeccionarlos uno a uno.

```python
import numpy as np
import sympy as sym
from IPython.display import Math, display

A = np.ones((3, 2))  # armamos una matriz de 3x2 con 1 en todas las posiciones
print("A=")  # la mostramos
display(Math(sym.latex(sym.sympify(A))))
print()
```

```python
A[0:1] = 3*A[0:1] #multiplicamos la fila 1 por 3
print("A=")
display(Math(sym.latex(sym.sympify(A)))) # la mostramos
```

```python
A_t = A.T # esta es la matriz traspuesta 
print("A_T=")
display(Math(sym.latex(sym.sympify(A_t)))) # mostramos la matriz traspuesta

print("A=")
display(Math(sym.latex(sym.sympify(A))))
```

```python
A[:, 1] = A[:, 1] * 4 # multiplicamos por 4 la segunda columna
print("A=")
display(Math(sym.latex(sym.sympify(A)))) # mostramos la matriz resultante
```

# <font color="darkcyan">¿Cómo se presenta un DataFrame?</font>

El formato de estas estructuras puede compararse con los diccionarios de Python. Efectivamente, las claves son los nombres de las columnas y los valores son las Series. Su estructura puede considerarse similar a la de una hoja de cálculo de Excel.

Cada fila contiene datos específicos de varias columnas, que son variables. El nombre de las filas de un DataFrame se llama «index» que, por defecto, empieza siempre por 0.

Para poder trabajar con un DataFrame importamos la librería pandas. 
Pandas es una de las principales librerías de Python. Las estructuras más utilizadas en este módulo son  los DataFrames.

```python
import matplotlib.pyplot as plt
import numpy as np

# Se puede cambiar el intervalo horizontal con xlim
# Se puede cambiar el intervalo vertical con ylim

x = np.linspace(0, 5, 30)  # arma 30 puntos entre el 0 y el 5
x2 = x * x
x3 = x**3
plt.xlim(0, 5)  # nuevos límites para el eje OX

plt.plot(
    x, x, "b.", x, x2, "rd", x, x3, "g^"
)  # representamos las tres funciones en el mismo diagrama
plt.gca().legend(("Lineal", "Cuadrática", "Cúbica"))  # agregamos una referencia

plt.show()
```

```python
# Para mostrar múltiples gráficos

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 6.28, 1000)  # armamos un vector con 1000 puntos entre 0 y 6.28

f = plt.figure(figsize=(10, 3))  # definimos el tamaño de la figura

ax = f.add_subplot(121)  # generamos un primer sub-gráfico
ax2 = f.add_subplot(122)  # generamos un segundo sub-gráfico

ax.plot(
    x, np.sin(x), "g:"
)  # representamos en verde un ciclo de la función seno con línea punteada
ax2.plot(
    x, np.cos(x), "r:"
)  # representamos en rojo un ciclo de la función coseno con línea punteada

plt.show()
```

### Varias funciones superpuestas

```python
x = np.linspace(0, 2 * np.pi)
sin = np.sin(x)  # defino la función seno sobre el vector de puntos
cos = np.cos(x)  # defino la función coseno sobre el vector de puntos

fig, ax = plt.subplots()  # en el mismo gráfico superpongo ambas

ax.plot(x, sin, "o")


ax.plot(x, cos, "*")
```

### Agregamos leyendas para distinguir las gráficas

```python
fig, ax = plt.subplots()

ax.plot(x, sin, label="sin")

ax.plot(x, cos, label="cos")

ax.legend()
```

## Podemos incluir símbolos matemáticos

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

ax.legend()
```

### Ubicación de la leyenda

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

ax.legend()
ax.legend(loc="upper right")
```

### Aplicando estilos al gráfico

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

sin_style = dict(linewidth=3, color="darkorange")
cos_style = dict(marker="o", markerfacecolor="limegreen", color="darkgreen")

ax.plot(x, sin, label="$f_1(x) = sin(x)$", **sin_style)

ax.plot(x, cos, label="$f_2(x) = cos(x)$", **cos_style)
```
