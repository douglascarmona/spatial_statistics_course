# <font color='darkgreen'> Sets o Conjuntos </font>

Los sets en Python son una estructura de datos usada para almacenar elementos de una manera similar a las listas, pero con algunas diferencias.

Debido a que los conjuntos no pueden tener m√∫ltiples ocurrencias del mismo elemento, hace que los conjuntos sean muy √∫tiles para eliminar eficientemente los valores duplicados de una lista o tuple y realizar operaciones de matem√°ticas comunes como sindicatos e intersecciones.

Para crear un set en Python se puede hacer con <font color='darkred'>Sets o Conjuntos</font>
 y pasando como entrada cualquier tipo iterable, como puede ser una lista. Se puede ver como a pesar de pasar elementos duplicados como dos 8 y en un orden determinado, al imprimir el set no conserva ese orden y los duplicados se han eliminado.

```python
# Crear set Python

s = set([5, 4, 6, 8, 8, 1])
print(s)       
print(type(s)) 
```

### Operaciones con sets


A diferencia de las listas, en los set no podemos modificar un elemento a trav√©s de su √≠ndice. Si lo intentamos, tendremos un TypeError.

```python
s = set([5, 6, 7, 8]) # declaro el set
#s[0] = 3 # intento modificar el primer elemento
```

Los elementos de un set deben ser **inmutables**, por lo que un elemento de un set no puede ser ni un diccionario ni una lista. Si lo intentamos tendremos un TypeError

```python
lista = ["Per√∫", "Argentina"]
#s = set(["M√©xico", "Espa√±a", lista])
```

**Los sets se pueden iterar de la misma forma que las listas.**

```python
s = set([5, 6, 7, 8])
for i in s:
    print(i) 
```

    
Con la funci√≥n **len()** podemos saber <font color='darkcyan'>la longitud total del set</font>. Como ya hemos indicado, los duplicados son eliminados.



```python
s = set([1, 2, 2, 3, 4])
print(len(s)) 
```

Tambi√©n podemos saber si un elemento est√° presente en un set con el operador in. Si el valor existe en el set, se devolver√° True.

```python
s = set(["Guitarra", "Bajo"])
print("Guitarra" in s) 
print("Piano" in s)
```

## <font color='darkgreen'> M√©todos sets</font>



Los sets tienen adem√°s diferentes funcionalidades, que se pueden aplicar en forma de operador o de m√©todo. Por ejemplo, el operador | nos permite realizar la uni√≥n de dos sets, lo que equivale a juntarlos. El equivalente es el m√©todo union() que vemos a continuaci√≥n.

```python
s1 = set([1, 2, 3]) # declaramos el primer set
s2 = set([3, 4, 5]) # declaramos el segundo set
print(s1 | s2) # unimos los dos sets y lo mostramos
```

**El m√©todo add() permite a√±adir un elemento al set.**

s.add(<elem>)

```python
set1 = set([1, 2])
set1.add(3)
print(set1)
```

**El m√©todo remove() elimina el elemento que se pasa como par√°metro. 
Si no se encuentra, se lanza la excepci√≥n KeyError.**

 
s.remove(<elem>)

```python
set2 = set([1, 2,4,7])
set2.remove(2)
print(set2) 
```

**El m√©todo discard() es muy parecido al remove(), borra el elemento que se pasa como par√°metro,y si no se encuentra no hace nada.**

s.discard(<elem>)

```python
set3 = set([1, 2, 5])
set3.discard(3)
print(set3)
```

**El m√©todo pop() elimina un elemento aleatorio del set.**

s.pop()

```python
set4 = set([1, 2, 9, "A"])
set4.pop()
print(set4) 
```

**El m√©todo clear() elimina todos los elementos de set.**

s.clear()

```python
set5 = set([1, 8, 6, "B"])
set5.clear()
print(set5)
```

Los sets cuentan con una gran cantidad de m√©todos que permiten realizar operaciones con dos o m√°s, como la uni√≥n o la intersecci√≥n.

Podemos calcular **la uni√≥n entre dos sets usando el m√©todo union()**. Esta operaci√≥n representa la ‚Äúmezcla‚Äù de ambos sets. N√≥tese que el m√©todo puede ser llamado con m√°s par√°metros de entrada, y su resultado ser√° la uni√≥n de todos los sets

```python

s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.union(s2)) 
```

Tambi√©n podemos calcular **la intersecci√≥n entre dos o m√°s set**. Su resultado ser√°n aquellos elementos que pertenecen a ambos sets.

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.intersection(s2)) 
```

# <font color="darkgreen">Slicing</font> 

La expresi√≥n slicing hace referencia a la operaci√≥n por medio de la cual se extraen elementos de una secuencia, tal como una lista o una cadena de caracteres. Dependiendo del caso, los elementos podr√≠an ser consecutivos o podr√≠an estar separados dentro de la secuencia original. Dado que esta operaci√≥n es de uso bastante frecuente, Python ofrece caracter√≠sticas que hacen posible construir slices de forma breve, sin tener que construir funciones especializadas para hacerlo

Supongamos que tenemos la siguiente cadena de caracteres con la que inicia el himno de Argentina:

```python
cadena = '¬°Oid mortales el grito sagrado: Libertad,Libertad,Libertad!'
```

C√≥mo podemos hacer para extraer elementos de esta cadena usando par√©ntesis cuadrados y una posici√≥n numerada desde cero.

```python
cadena[0] # estamos extrayendo un slice de la cadena original con el primer caracter
```

```python
cadena[1:4] # extraemos los caracteres desde la posici√≥n 1 hasta la 4 (excluida), 
```

```python
cadena[30:] # recortamos de la posici√≥n 30 en adelante
```

```python
cadena[:30] # recortamos hasta la posici√≥n 30 exclusive
```

```python
cadena[None:30] # otra forma de hacer lo mismo que la anterior
```

```python
# Crear una copia de la cadena original
cadena2 = cadena[:]
cadena2
```

### <span style="color: darkblue;">Para acceder a los elementos de una matriz o recorrerlos, podemos tambi√©n utilizar funciones l√≥gicas y obtener resultados "booleanos" (True or False). <span/>
Por ejemplo, podr√≠amos preguntamos que elementos de la matriz son mayores a cero e inspeccionarlos uno a uno.

```python
import numpy as np
import sympy as sym
from IPython.display import Math, display

A = np.ones((3, 2))  # armamos una matriz de 3x2 con 1 en todas las posiciones
print("A=")  # la mostramos
display(Math(sym.latex(sym.sympify(A))))
print()
```

```python
A[0:1] = 3*A[0:1] #multiplicamos la fila 1 por 3
print("A=")
display(Math(sym.latex(sym.sympify(A)))) # la mostramos
```

```python
A_t = A.T # esta es la matriz traspuesta 
print("A_T=")
display(Math(sym.latex(sym.sympify(A_t)))) # mostramos la matriz traspuesta

print("A=")
display(Math(sym.latex(sym.sympify(A))))
```

```python
A[:, 1] = A[:, 1] * 4 # multiplicamos por 4 la segunda columna
print("A=")
display(Math(sym.latex(sym.sympify(A)))) # mostramos la matriz resultante
```

El formato de estas estructuras puede compararse con los diccionarios de Python. Efectivamente, las claves son los nombres de las columnas y los valores son las Series. Su estructura puede considerarse similar a la de una hoja de c√°lculo de Excel.

Cada fila contiene datos espec√≠ficos de varias columnas, que son variables. El nombre de las filas de un DataFrame se llama ¬´index¬ª que, por defecto, empieza siempre por 0.

Para poder trabajar con un DataFrame importamos la librer√≠a pandas. 
Pandas es una de las principales librer√≠as de Python. Las estructuras m√°s utilizadas en este m√≥dulo son  los DataFrames.

```python
import matplotlib.pyplot as plt
import numpy as np

# Se puede cambiar el intervalo horizontal con xlim
# Se puede cambiar el intervalo vertical con ylim

x = np.linspace(0, 5, 30)  # arma 30 puntos entre el 0 y el 5
x2 = x * x
x3 = x**3
plt.xlim(0, 5)  # nuevos l√≠mites para el eje OX

plt.plot(
    x, x, "b.", x, x2, "rd", x, x3, "g^"
)  # representamos las tres funciones en el mismo diagrama
plt.gca().legend(("Lineal", "Cuadr√°tica", "C√∫bica"))  # agregamos una referencia

plt.show()
```

```python
# Para mostrar m√∫ltiples gr√°ficos

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 6.28, 1000)  # armamos un vector con 1000 puntos entre 0 y 6.28

f = plt.figure(figsize=(10, 3))  # definimos el tama√±o de la figura

ax = f.add_subplot(121)  # generamos un primer sub-gr√°fico
ax2 = f.add_subplot(122)  # generamos un segundo sub-gr√°fico

ax.plot(
    x, np.sin(x), "g:"
)  # representamos en verde un ciclo de la funci√≥n seno con l√≠nea punteada
ax2.plot(
    x, np.cos(x), "r:"
)  # representamos en rojo un ciclo de la funci√≥n coseno con l√≠nea punteada

plt.show()
```

### Varias funciones superpuestas

```python
x = np.linspace(0, 2 * np.pi)
sin = np.sin(x)  # defino la funci√≥n seno sobre el vector de puntos
cos = np.cos(x)  # defino la funci√≥n coseno sobre el vector de puntos

fig, ax = plt.subplots()  # en el mismo gr√°fico superpongo ambas

ax.plot(x, sin, "o")


ax.plot(x, cos, "*")
```

### Agregamos leyendas para distinguir las gr√°ficas

```python
fig, ax = plt.subplots()

ax.plot(x, sin, label="sin")

ax.plot(x, cos, label="cos")

ax.legend()
```

## Podemos incluir s√≠mbolos matem√°ticos

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

ax.legend()
```

### Ubicaci√≥n de la leyenda

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

ax.legend()
ax.legend(loc="upper right")
```

### Aplicando estilos al gr√°fico

```python
fig, ax = plt.subplots()
ax.plot(x, sin, label="$f_1(x) = sin(x)$")

ax.plot(x, cos, label="$f_2(x) = cos(x)$")

sin_style = dict(linewidth=3, color="darkorange")
cos_style = dict(marker="o", markerfacecolor="limegreen", color="darkgreen")

ax.plot(x, sin, label="$f_1(x) = sin(x)$", **sin_style)

ax.plot(x, cos, label="$f_2(x) = cos(x)$", **cos_style)
```

### <font color="#229966">Histograma por grupo</font>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el grid de Facet para las clases
g = sns.FacetGrid(titanic, col="pclass", height=5, aspect=1)

# Definir la paleta de colores
paleta = sns.color_palette("viridis", n_colors=len(titanic["pclass"].unique()))

# Mapear el histograma a cada subplot
for i, ax in enumerate(g.axes.flat):
    ax.hist(
        titanic.loc[titanic["pclass"] == i + 1, "age"],
        bins=10,
        color=paleta[i],
        edgecolor="black",
    )

# Agregar t√≠tulos y etiquetas
g.set_axis_labels("Edad", "Frecuencia")
g.set_titles("Clase {col_name}")
g.fig.suptitle("Histograma de Edades por Clase", y=1.05)
plt.show()
```

### <font color="#229966">Densidad por grupo</font>

```python
# Crear el grid de Facet para las clases
order = sorted(titanic["pclass"].unique())
g = sns.FacetGrid(titanic, col="pclass", height=5, aspect=1)
g.map(sns.kdeplot, "age", fill=True, color="darkcyan")

# Agregar t√≠tulos y etiquetas
g.set_axis_labels("Edad", "Densidad")
g.set_titles("Clase {col_name}")
g.fig.suptitle("Densidad de Edades por Clase", y=1.05)
plt.show()
```
